```python
from huggingface_hub import HfApi
import arxiv
import pandas as pd

api = HfApi()
models = api.list_models(filter="text-classification", full=True, cardData=True, fetch_config=True)
```

```python
data_list = []
i = 0
for model in models:  
    try:
        arxiv_ids = [x for x in model.tags if x.startswith("arxiv")]
        arxiv_ids_str = ", ".join(arxiv_ids) if arxiv_ids else None

        base_model = model.card_data.get("base_model", None) if model.card_data else None
        if isinstance(base_model, list):
            base_model = ", ".join(base_model)
        elif not isinstance(base_model, str):
            base_model = None

        language = model.card_data.get("language", None) if model.card_data else None
        if isinstance(language, list):
            language = ", ".join(language)  
        elif not isinstance(language, str):
            language = None  
        data_list.append([
            model.id,
            model.author,
            model.created_at,
            model.downloads,
            model.pipeline_tag,
            arxiv_ids_str,  
            base_model,
            language  
        ])
        
        i += 1
        if i == 5000:
            break

    except ValueError as e:
        print(f"Skipping model {model.id} due to error: {e}")

# Create DataFrame
df = pd.DataFrame(data_list, columns=['id', 'author', 'created_at', 'downloads', 'pipeline_tag', 'arxiv_ids', 'base_model', 'language'])

# Display DataFrame
display(df)

# Save to CSV
df.to_csv("./output/models.csv", index=False)
```

```python
i = 0
evaluation_results = []

for model in models:
    try:
        if hasattr(model.card_data, 'eval_results') and model.card_data.eval_results:
            for result in model.card_data.eval_results:
                evaluation_results.append([
                    model.id,
                    result.task_type,
                    result.dataset_type,
                    result.dataset_name,
                    result.metric_type,
                    result.metric_value
                ])
            i += 1
            if i == 500:
                break
    except Exception as e:
        print(f"Skipping model {model.id} due to error: {e}")

df_eval = pd.DataFrame(evaluation_results, columns=['id', 'task_type', 'dataset_type', 'dataset_name', 'metric_type', 'metric_value'])

# Display DataFrame
display(df_eval)

# Save to CSV
df_eval.to_csv("./output/evaluation_results.csv", index=False)
```

```python
import pandas as pd
import json
from huggingface_hub import HfApi

api = HfApi()
models = api.list_models(filter="text-classification", full=True, cardData=True, fetch_config=True)

evaluation_results = []

for model in models:
    try:
        if hasattr(model.card_data, 'eval_results') and model.card_data.eval_results:
            eval_list = []  # Store all evaluation results for a model
            
            for result in model.card_data.eval_results:
                eval_list.append({
                    "task_type": result.task_type,
                    "dataset_type": result.dataset_type,
                    "dataset_name": result.dataset_name,
                    "metric_type": result.metric_type,
                    "metric_value": result.metric_value
                })

            # Append model ID and all evaluation results as a JSON string
            evaluation_results.append({
                'id': model.id,
                'evaluation_metrics': json.dumps(eval_list)  # Store as a JSON string
            })

        # Stop at 500 models with evaluations
        if len(evaluation_results) >= 500:
            break

    except Exception as e:
        print(f"Skipping model {model.id} due to error: {e}")

# Convert evaluation results into a DataFrame
df_eval = pd.DataFrame(evaluation_results)

# Display the final DataFrame
display(df_eval)

# Save to CSV
df_eval.to_csv("./output/evaluation_results.csv", index=False)

```
